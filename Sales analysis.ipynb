<H2> Exploratory Data Analysis

- <h5> What is the overall sales trend? 

```python
data['month_year'] = data['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
df_monthly_trend = data.groupby('month_year').sum()['sales'].reset_index()
df_monthly_trend
```

```python
plt.figure(figsize = (15,6))
plt.plot(df_monthly_trend['month_year'], df_monthly_trend['sales'], color ='green')
plt.xticks(rotation = 'vertical', size=8)
plt.show()
```

- <h5> Which are the Top 10 products by sales?

```python
pro_sales = pd.DataFrame(data.groupby('product_name').sum()['sales'])
```

```python
pro_sales = pro_sales.sort_values('sales', ascending=False)
```

```python
pro_sales[:10]
```

- <h5> Which are the Most Selling Products?

```python
most_sold_products= pd.DataFrame(data.groupby('product_name').sum()['quantity'])
```

```python
most_sold_products= most_sold_products.sort_values('quantity',ascending=False)
```

```python
most_sold_products[:10]
```

- <h5> Which is the most preferred Ship Mode?

```python
plt.figure(figsize=(9,8.5))
sns.countplot(data['ship_mode'])
plt.show()
```

- <h5> Which are the Most Profitable Category and Sub-Category?

```python
profit_category = pd.DataFrame(data.groupby(['category','sub_category']).sum()['profit'])
```

```python
profit_category
```

```python
profit_category.sort_values(['category','profit'], ascending=False)
```
